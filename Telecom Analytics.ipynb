### Customer Risk Score using function and if-else

#### Complaint count per customer

```python
complaint_count = network_issue.groupby("customer_id")["issue_id"].count().reset_index()

complaint_count.rename(columns={"issue_id":"total_complaints"}, inplace=True)

complaint_count.head()
```

#### Average Resolution time

```python
avg_resolution = network_issue.groupby("customer_id")["resolution_time_hrs"].mean().reset_index()

avg_resolution.rename(columns={"resolution_time_hrs":"avg_resolution_time"}, inplace=True)

avg_resolution.head()
```

#### Payment Delayed or Unpaid

```python
billing["payment_risk"] = billing["payment_status"].apply(
    lambda x: 1 if x != "Paid" else 0
)

payment_risk = billing.groupby("customer_id")["payment_risk"].sum().reset_index()

payment_risk.head()
```

#### Heavy Usage per Customer

```python
usage_risk = network_usage.groupby("customer_id")["heavy_usage_flag"].max().reset_index()

usage_risk.head()
```

#### Combine Customer Risk Inputs

```python
customer_risk = complaint_count.merge(avg_resolution, on="customer_id", how="left")

customer_risk = customer_risk.merge(payment_risk, on="customer_id", how="left")

customer_risk = customer_risk.merge(usage_risk, on="customer_id", how="left")
```

```python
customer_risk.fillna(0, inplace=True)
```

#### Customer Risk Score Function

```python
def calculate_risk_score(row):

    score = 0
    if row["total_complaints"] > 5:
        score += 2
    
    if row["avg_resolution_time"] > 48:
        score += 2
        
    if row["payment_risk"] > 0:
        score += 2
        
    if row["heavy_usage_flag"] == 1:
        score += 1
        
    return score
```

```python
customer_risk["risk_score"] = customer_risk.apply(calculate_risk_score, axis=1)
```

#### Customer Risk Category

```python
def risk_category(score):
    
    if score >= 5:
        return "High Risk"
    
    elif score >= 3:
        return "Medium Risk"
    
    else:
        return "Low Risk"


customer_risk["risk_category"] = customer_risk["risk_score"].apply(risk_category)
```
